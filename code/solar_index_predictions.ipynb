```python
# ============================================================================
# SOLAR INDEX PREDICTION FOR MUMBAI - COMPLETE PROJECT CODE
# Time Series Analysis and Machine Learning Approach
# ============================================================================

print("""
****************************************************************
        SOLAR INDEX PREDICTION FOR MUMBAI REGIONS              
        Using Time Series Analysis & Machine Learning           
****************************************************************
""")

# ==================== STEP 1: INSTALL & IMPORT LIBRARIES ====================
print("\n" + "="*70)
print("INSTALLING & IMPORTING LIBRARIES")
print("="*70)

import subprocess
import sys

# Install required packages
packages = ['pandas', 'numpy', 'matplotlib', 'seaborn', 'scikit-learn', 
            'xgboost', 'statsmodels', 'prophet', 'requests']

print("\nInstalling packages (this may take 2-3 minutes)...")
for package in packages:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package])

print("All packages installed!\n")

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import requests
import warnings
import os
warnings.filterwarnings('ignore')

# Time series and ML imports
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from prophet import Prophet

print("All libraries imported successfully!\n")

# ==================== STEP 2: DATA COLLECTION ====================
print("="*70)
print("FETCHING REAL-TIME SOLAR INDEX DATA FROM NASA POWER API")
print("="*70)

def fetch_solar_index_data(latitude, longitude, location_name, years=3):
    """
    Fetch Solar Index (Solar Irradiance) data from NASA POWER API
    
    Solar Index Parameters:
    - ALLSKY_SFC_SW_DWN: All-sky surface shortwave downward irradiance (kWh/m²/day)
    - This represents the Solar Index for renewable energy applications
    """
    
    end_date = datetime.now().strftime('%Y-%m-%d')
    start_date = (datetime.now() - timedelta(days=years*365)).strftime('%Y-%m-%d')
    
    base_url = "https://power.larc.nasa.gov/api/temporal/daily/point"
    
    parameters = {
        'parameters': 'ALLSKY_SFC_SW_DWN,T2M,RH2M,WS2M,PRECTOTCORR',
        'community': 'RE',
        'longitude': longitude,
        'latitude': latitude,
        'start': start_date.replace('-', ''),
        'end': end_date.replace('-', ''),
        'format': 'JSON'
    }
    
    print(f"Location: {location_name}")
    print(f"   Coordinates: ({latitude}N, {longitude}E)")
    print(f"   Date Range: {start_date} to {end_date}")
    print(f"   Fetching Solar Index data...")
    
    try:
        response = requests.get(base_url, params=parameters, timeout=60)
        
        if response.status_code == 200:
            data = response.json()
            parameters_data = data['properties']['parameter']
            
            df = pd.DataFrame({
                'Date': pd.to_datetime(list(parameters_data['ALLSKY_SFC_SW_DWN'].keys()), 
                                      format='%Y%m%d'),
                'Solar_Index': list(parameters_data['ALLSKY_SFC_SW_DWN'].values()),
                'Temperature': list(parameters_data['T2M'].values()),
                'Humidity': list(parameters_data['RH2M'].values()),
                'Wind_Speed': list(parameters_data['WS2M'].values()),
                'Precipitation': list(parameters_data['PRECTOTCORR'].values())
            })
            
            df['Location'] = location_name
            print(f"Successfully fetched {len(df)} days of Solar Index data!")
            print(f"   Parameter: All-Sky Surface Shortwave Downward Irradiance")
            print(f"   Unit: kWh/m²/day")
            
            return df
        else:
            print(f"API Error: {response.status_code}")
            print(f"   Message: {response.text}")
            return None
            
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None

# Mumbai locations
locations = {
    'Ghatkopar': {'lat': 19.0860, 'lon': 72.9081},
    'South_Mumbai': {'lat': 18.9220, 'lon': 72.8347}
}

# Select primary location
LOCATION = 'Ghatkopar'
print(f"Primary Location: {LOCATION}, Mumbai")

# Fetch data
df = fetch_solar_index_data(
    locations[LOCATION]['lat'],
    locations[LOCATION]['lon'],
    LOCATION,
    years=3
)

if df is None:
    print("Failed to fetch data. Please check your internet connection.")
    sys.exit()

# Save raw data
df.to_csv(f'solar_index_data_{LOCATION}.csv', index=False)
print(f"Raw data saved as: 'solar_index_data_{LOCATION}.csv'")

# ==================== STEP 3: DATA PREPROCESSING ====================
print("\n" + "="*70)
print("DATA PREPROCESSING & FEATURE ENGINEERING")
print("="*70)

# Data quality check
print("\nData Quality Check:")
print(f"   Total records: {len(df)}")
print(f"   Date range: {df['Date'].min()} to {df['Date'].max()}")
print(f"   Missing values: {df.isnull().sum().sum()}")

# Handle missing values
df = df.fillna(method='ffill').fillna(method='bfill')

# Remove invalid values (NASA uses -999 for missing)
df = df[df['Solar_Index'] >= 0]
print(f"   Valid records after cleaning: {len(df)}")

# Feature Engineering
print("\nCreating Features:")

# Temporal features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['DayOfYear'] = df['Date'].dt.dayofyear
df['Week'] = df['Date'].dt.isocalendar().week
df['Quarter'] = df['Date'].dt.quarter

print("   Temporal features (Year, Month, Day, Week, Quarter)")

# Seasonal classification
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5, 6]:
        return 'Summer'
    elif month in [7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

df['Season'] = df['Month'].apply(get_season)
print("   Season classification")

# Lag features (previous values)
for lag in [1, 2, 3, 7, 14, 30]:
    df[f'Solar_Lag_{lag}'] = df['Solar_Index'].shift(lag)

print("   Lag features (1, 2, 3, 7, 14, 30 days)")

# Rolling statistics
for window in [3, 7, 14, 30]:
    df[f'Solar_Rolling_Mean_{window}'] = df['Solar_Index'].rolling(window=window).mean()
    df[f'Solar_Rolling_Std_{window}'] = df['Solar_Index'].rolling(window=window).std()
    df[f'Solar_Rolling_Min_{window}'] = df['Solar_Index'].rolling(window=window).min()
    df[f'Solar_Rolling_Max_{window}'] = df['Solar_Index'].rolling(window=window).max()

print("   Rolling statistics (3, 7, 14, 30 days)")

# Weather interaction features
df['Temp_Humidity_Interaction'] = df['Temperature'] * df['Humidity']
df['Wind_Precip_Interaction'] = df['Wind_Speed'] * df['Precipitation']

print("   Weather interaction features")

# Exponential weighted moving average
df['Solar_EWMA_7'] = df['Solar_Index'].ewm(span=7).mean()
df['Solar_EWMA_30'] = df['Solar_Index'].ewm(span=30).mean()

print("   Exponential weighted moving averages")

# Drop rows with NaN (from lag/rolling features)
df_clean = df.dropna().reset_index(drop=True)

print(f"\nPreprocessing complete!")
print(f"   Final dataset: {len(df_clean)} records")
print(f"   Total features: {len(df_clean.columns)} columns")

# Save processed data
df_clean.to_csv(f'solar_index_data_{LOCATION}_processed.csv', index=False)
print(f"   Saved processed data: 'solar_index_data_{LOCATION}_processed.csv'")

# Statistical summary
print("\nSolar Index Statistics:")
print(f"   Mean:   {df_clean['Solar_Index'].mean():.2f} kWh/m²/day")
print(f"   Median: {df_clean['Solar_Index'].median():.2f} kWh/m²/day")
print(f"   Std:    {df_clean['Solar_Index'].std():.2f} kWh/m²/day")
print(f"   Min:    {df_clean['Solar_Index'].min():.2f} kWh/m²/day")
print(f"   Max:    {df_clean['Solar_Index'].max():.2f} kWh/m²/day")

# ==================== STEP 4: EXPLORATORY DATA ANALYSIS ====================
print("\n" + "="*70)
print("EXPLORATORY DATA ANALYSIS")
print("="*70)

# Create visualizations directory
os.makedirs('visualizations', exist_ok=True)

# Set plot style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("\nCreating visualizations...")

# Visualization 1: Time Series Overview
print("   [1/10] Time series overview...")
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

axes[0].plot(df_clean['Date'], df_clean['Solar_Index'], linewidth=1, alpha=0.8, color='orange')
axes[0].set_title(f'Solar Index Time Series - {LOCATION}, Mumbai', fontsize=16, fontweight='bold')
axes[0].set_ylabel('Solar Index (kWh/m²/day)', fontsize=12)
axes[0].grid(True, alpha=0.3)

# With rolling average
axes[1].plot(df_clean['Date'], df_clean['Solar_Index'], linewidth=1, alpha=0.5, 
            label='Daily Solar Index', color='orange')
axes[1].plot(df_clean['Date'], df_clean['Solar_Rolling_Mean_30'], linewidth=2, 
            label='30-Day Moving Average', color='red')
axes[1].set_title('Solar Index with 30-Day Moving Average', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Date', fontsize=12)
axes[1].set_ylabel('Solar Index (kWh/m²/day)', fontsize=12)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('visualizations/1_solar_index_timeseries.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 2: Seasonal Analysis
print("   [2/10] Seasonal patterns...")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Monthly average
monthly_avg = df_clean.groupby('Month')['Solar_Index'].agg(['mean', 'std'])
axes[0, 0].bar(monthly_avg.index, monthly_avg['mean'], color='coral', alpha=0.7, 
              yerr=monthly_avg['std'], capsize=5)
axes[0, 0].set_title('Average Solar Index by Month', fontweight='bold', fontsize=14)
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Solar Index (kWh/m²/day)')
axes[0, 0].set_xticks(range(1, 13))
axes[0, 0].grid(axis='y', alpha=0.3)

# Seasonal average
season_order = ['Winter', 'Summer', 'Monsoon', 'Post-Monsoon']
seasonal_data = df_clean.groupby('Season')['Solar_Index'].agg(['mean', 'std'])
seasonal_data = seasonal_data.reindex(season_order)
axes[0, 1].bar(seasonal_data.index, seasonal_data['mean'], color='skyblue', alpha=0.7,
              yerr=seasonal_data['std'], capsize=5)
axes[0, 1].set_title('Average Solar Index by Season', fontweight='bold', fontsize=14)
axes[0, 1].set_xlabel('Season')
axes[0, 1].set_ylabel('Solar Index (kWh/m²/day)')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].grid(axis='y', alpha=0.3)

# Distribution
axes[1, 0].hist(df_clean['Solar_Index'], bins=50, color='gold', edgecolor='black', alpha=0.7)
axes[1, 0].axvline(df_clean['Solar_Index'].mean(), color='red', linestyle='--', 
                   linewidth=2, label=f'Mean: {df_clean["Solar_Index"].mean():.2f}')
axes[1, 0].set_title('Solar Index Distribution', fontweight='bold', fontsize=14)
axes[1, 0].set_xlabel('Solar Index (kWh/m²/day)')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].legend()
axes[1, 0].grid(axis='y', alpha=0.3)

# Box plot by season
season_data_full = [df_clean[df_clean['Season'] == s]['Solar_Index'].values 
                    for s in season_order]
bp = axes[1, 1].boxplot(season_data_full, labels=season_order, patch_artist=True)
for patch in bp['boxes']:
    patch.set_facecolor('lightgreen')
axes[1, 1].set_title('Solar Index Distribution by Season', fontweight='bold', fontsize=14)
axes[1, 1].set_xlabel('Season')
axes[1, 1].set_ylabel('Solar Index (kWh/m²/day)')
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('visualizations/2_seasonal_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 3: Correlation Analysis
print("   [3/10] Correlation matrix...")
corr_features = ['Solar_Index', 'Temperature', 'Humidity', 'Wind_Speed', 'Precipitation',
                'Solar_Lag_1', 'Solar_Lag_7', 'Solar_Rolling_Mean_7']
corr_matrix = df_clean[corr_features].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', center=0, fmt='.2f',
           square=True, linewidths=1, cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Feature Correlation Matrix - Solar Index Prediction', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('visualizations/3_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 4: Year-over-Year Comparison
print("   [4/10] Year-over-year comparison...")
fig, ax = plt.subplots(figsize=(16, 6))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
for idx, year in enumerate(sorted(df_clean['Year'].unique())):
    year_data = df_clean[df_clean['Year'] == year]
    ax.plot(year_data['DayOfYear'], year_data['Solar_Index'],
           label=f'{year}', linewidth=2, alpha=0.7, color=colors[idx % len(colors)])
ax.set_title('Solar Index - Year-over-Year Comparison', fontsize=16, fontweight='bold')
ax.set_xlabel('Day of Year', fontsize=12)
ax.set_ylabel('Solar Index (kWh/m²/day)', fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/4_year_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 5: Weather Impact
print("   [5/10] Weather impact analysis...")
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Temperature vs Solar Index
axes[0, 0].scatter(df_clean['Temperature'], df_clean['Solar_Index'], 
                  alpha=0.3, s=20, c=df_clean['Solar_Index'], cmap='YlOrRd')
axes[0, 0].set_title('Solar Index vs Temperature', fontweight='bold', fontsize=14)
axes[0, 0].set_xlabel('Temperature (C)')
axes[0, 0].set_ylabel('Solar Index (kWh/m²/day)')
axes[0, 0].grid(True, alpha=0.3)

# Humidity vs Solar Index
axes[0, 1].scatter(df_clean['Humidity'], df_clean['Solar_Index'],
                  alpha=0.3, s=20, c=df_clean['Solar_Index'], cmap='YlOrRd')
axes[0, 1].set_title('Solar Index vs Humidity', fontweight='bold', fontsize=14)
axes[0, 1].set_xlabel('Humidity (%)')
axes[0, 1].set_ylabel('Solar Index (kWh/m²/day)')
axes[0, 1].grid(True, alpha=0.3)

# Wind Speed vs Solar Index
axes[1, 0].scatter(df_clean['Wind_Speed'], df_clean['Solar_Index'],
                  alpha=0.3, s=20, c=df_clean['Solar_Index'], cmap='YlOrRd')
axes[1, 0].set_title('Solar Index vs Wind Speed', fontweight='bold', fontsize=14)
axes[1, 0].set_xlabel('Wind Speed (m/s)')
axes[1, 0].set_ylabel('Solar Index (kWh/m²/day)')
axes[1, 0].grid(True, alpha=0.3)

# Precipitation vs Solar Index
axes[1, 1].scatter(df_clean['Precipitation'], df_clean['Solar_Index'],
                  alpha=0.3, s=20, c=df_clean['Solar_Index'], cmap='YlOrRd')
axes[1, 1].set_title('Solar Index vs Precipitation', fontweight='bold', fontsize=14)
axes[1, 1].set_xlabel('Precipitation (mm/day)')
axes[1, 1].set_ylabel('Solar Index (kWh/m²/day)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('visualizations/5_weather_impact.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 6: Monthly Heatmap
print("   [6/10] Monthly heatmap...")
pivot_data = df_clean.pivot_table(values='Solar_Index', index='Month', 
                                  columns='Year', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, annot=True, fmt='.2f', cmap='YlOrRd',
           cbar_kws={'label': 'Solar Index (kWh/m²/day)'})
plt.title('Average Solar Index by Month and Year', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.tight_layout()
plt.savefig('visualizations/6_monthly_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nVisualizations 1-6 completed!")

# ==================== STEP 5: TIME SERIES DECOMPOSITION ====================
print("\n" + "="*70)
print("TIME SERIES DECOMPOSITION")
print("="*70)

print("\nPerforming seasonal decomposition...")

# Prepare time series
ts_data = df_clean.set_index('Date')['Solar_Index']

# Decomposition
decomposition = seasonal_decompose(ts_data, model='additive', period=365)

# Plot decomposition
print("   [7/10] Time series decomposition...")
fig, axes = plt.subplots(4, 1, figsize=(16, 14))

axes[0].plot(decomposition.observed, color='blue', linewidth=1)
axes[0].set_ylabel('Observed', fontsize=11)
axes[0].set_title('Time Series Decomposition - Solar Index', fontsize=16, fontweight='bold')
axes[0].grid(True, alpha=0.3)

axes[1].plot(decomposition.trend, color='orange', linewidth=2)
axes[1].set_ylabel('Trend', fontsize=11)
axes[1].grid(True, alpha=0.3)

axes[2].plot(decomposition.seasonal, color='green', linewidth=1)
axes[2].set_ylabel('Seasonal', fontsize=11)
axes[2].grid(True, alpha=0.3)

axes[3].plot(decomposition.resid, color='red', linewidth=1, alpha=0.7)
axes[3].set_ylabel('Residual', fontsize=11)
axes[3].set_xlabel('Date', fontsize=12)
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('visualizations/7_decomposition.png', dpi=300, bbox_inches='tight')
plt.show()

# Stationarity test
adf_result = adfuller(ts_data.dropna())
print(f"\nAugmented Dickey-Fuller Test:")
print(f"   ADF Statistic: {adf_result[0]:.4f}")
print(f"   p-value: {adf_result[1]:.4f}")
print(f"   Critical Values:")
for key, value in adf_result[4].items():
    print(f"      {key}: {value:.3f}")
    
if adf_result[1] < 0.05:
    print(f"   Series is STATIONARY (p-value < 0.05)")
else:
    print(f"   Series is NON-STATIONARY (p-value >= 0.05)")

# ==================== STEP 6: MODEL PREPARATION ====================
print("\n" + "="*70)
print("PREPARING DATA FOR MODEL TRAINING")
print("="*70)

# Select features
feature_cols = [
    'Temperature', 'Humidity', 'Wind_Speed', 'Precipitation',
    'Month', 'DayOfYear', 'DayOfWeek',
    'Solar_Lag_1', 'Solar_Lag_2', 'Solar_Lag_3', 'Solar_Lag_7',
    'Solar_Rolling_Mean_3', 'Solar_Rolling_Mean_7', 'Solar_Rolling_Mean_14',
    'Solar_Rolling_Std_7', 'Solar_EWMA_7'
]

print(f"\nSelected Features ({len(feature_cols)}):")
for i, feat in enumerate(feature_cols, 1):
    print(f"   {i}. {feat}")

X = df_clean[feature_cols]
y = df_clean['Solar_Index']

# Train-test split (80-20)
split_idx = int(len(df_clean) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
dates_test = df_clean['Date'].iloc[split_idx:].values

print(f"\nData Split:")
print(f"   Training set: {len(X_train)} samples ({len(X_train)/len(df_clean)*100:.1f}%)")
print(f"   Testing set:  {len(X_test)} samples ({len(X_test)/len(df_clean)*100:.1f}%)")
print(f"   Train period: {df_clean['Date'].iloc[0]} to {df_clean['Date'].iloc[split_idx-1]}")
print(f"   Test period:  {df_clean['Date'].iloc[split_idx]} to {df_clean['Date'].iloc[-1]}")

# Feature scaling
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nFeatures scaled using MinMaxScaler")

# ==================== STEP 7: MODEL TRAINING ====================
print("\n" + "="*70)
print("TRAINING MACHINE LEARNING MODELS")
print("="*70)

results = {}

# MODEL 1: ARIMA
print("\n[1/4] Training ARIMA Model...")
print("   Configuration: ARIMA(5, 1, 2)")
try:
    arima_model = ARIMA(y_train, order=(5, 1, 2))
    arima_fit = arima_model.fit()
    arima_pred = arima_fit.forecast(steps=len(y_test))
    
    results['ARIMA'] = {
        'predictions': arima_pred.values,
        'mae': mean_absolute_error(y_test, arima_pred),
        'rmse': np.sqrt(mean_squared_error(y_test, arima_pred)),
        'r2': r2_score(y_test, arima_pred)
    }
    print(f"   Training complete!")
    print(f"      MAE:  {results['ARIMA']['mae']:.4f} kWh/m²/day")
    print(f"      RMSE: {results['ARIMA']['rmse']:.4f} kWh/m²/day")
    print(f"      R2:   {results['ARIMA']['r2']:.4f}")
except Exception as e:
    print(f"   ARIMA training failed: {e}")
    results['ARIMA'] = None

# MODEL 2: RANDOM FOREST
print("\n[2/4] Training Random Forest Model...")
print("   Configuration: 100 trees, max_depth=15")
rf_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train_scaled, y_train)
rf_pred = rf_model.predict(X_test_scaled)

results['Random Forest'] = {
    'predictions': rf_pred,
    'mae': mean_absolute_error(y_test, rf_pred),
    'rmse': np.sqrt(mean_squared_error(y_test, rf_pred)),
    'r2': r2_score(y_test, rf_pred),
    'model': rf_model
}
print(f"   Training complete!")
print(f"      MAE:  {results['Random Forest']['mae']:.4f} kWh/m²/day")
print(f"      RMSE: {results['Random Forest']['rmse']:.4f} kWh/m²/day")
print(f"      R2:   {results['Random Forest']['r2']:.4f}")

# MODEL 3: XGBOOST
print("\n[3/4] Training XGBoost Model...")
print("   Configuration: 100 estimators, max_depth=7, learning_rate=0.1")
xgb_model = XGBRegressor(
    n_estimators=100,
    max_depth=7,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    verbosity=0
)
xgb_model.fit(X_train_scaled, y_train)
xgb_pred = xgb_model.predict(X_test_scaled)

results['XGBoost'] = {
    'predictions': xgb_pred,
    'mae': mean_absolute_error(y_test, xgb_pred),
    'rmse': np.sqrt(mean_squared_error(y_test, xgb_pred)),
    'r2': r2_score(y_test, xgb_pred),
    'model': xgb_model
}
print(f"   Training complete!")
print(f"      MAE:  {results['XGBoost']['mae']:.4f} kWh/m²/day")
print(f"      RMSE: {results['XGBoost']['rmse']:.4f} kWh/m²/day")
print(f"      R2:   {results['XGBoost']['r2']:.4f}")

# MODEL 4: PROPHET
print("\n[4/4] Training Prophet Model...")
print("   Configuration: Yearly + Weekly seasonality")
try:
    prophet_df = df_clean[['Date', 'Solar_Index']].rename(
        columns={'Date': 'ds', 'Solar_Index': 'y'}
    )
    prophet_df_train = prophet_df[:split_idx]
    
    prophet_model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        seasonality_mode='multiplicative'
    )
    prophet_model.fit(prophet_df_train)
    
    future_dates = prophet_model.make_future_dataframe(periods=len(y_test))
    prophet_forecast = prophet_model.predict(future_dates)
    prophet_pred = prophet_forecast['yhat'].iloc[-len(y_test):].values
    
    results['Prophet'] = {
        'predictions': prophet_pred,
        'mae': mean_absolute_error(y_test, prophet_pred),
        'rmse': np.sqrt(mean_squared_error(y_test, prophet_pred)),
        'r2': r2_score(y_test, prophet_pred)
    }
    print(f"   Training complete!")
    print(f"      MAE:  {results['Prophet']['mae']:.4f} kWh/m²/day")
    print(f"      RMSE: {results['Prophet']['rmse']:.4f} kWh/m²/day")
    print(f"      R2:   {results['Prophet']['r2']:.4f}")
except Exception as e:
    print(f"   Prophet training failed: {e}")
    results['Prophet'] = None

# ==================== STEP 8: MODEL COMPARISON ====================
print("\n" + "="*70)
print("MODEL PERFORMANCE COMPARISON")
print("="*70)

# Create comparison table
comparison_data = []
for model_name in ['ARIMA', 'Random Forest', 'XGBoost', 'Prophet']:
    if results[model_name] is not None:
        comparison_data.append({
            'Model': model_name,
            'MAE (kWh/m²/day)': results[model_name]['mae'],
            'RMSE (kWh/m²/day)': results[model_name]['rmse'],
            'R2 Score': results[model_name]['r2'],
            'Accuracy (%)': results[model_name]['r2'] * 100
        })

comparison_df = pd.DataFrame(comparison_data)
comparison_df = comparison_df.sort_values('MAE (kWh/m²/day)')

print("\n" + "="*70)
print(comparison_df.to_string(index=False))
print("="*70)

# Find best model
best_idx = comparison_df['MAE (kWh/m²/day)'].idxmin()
best_model_name = comparison_df.loc[best_idx, 'Model']
best_mae = comparison_df.loc[best_idx, 'MAE (kWh/m²/day)']
best_r2 = comparison_df.loc[best_idx, 'R2 Score']

print(f"\nBEST MODEL: {best_model_name}")
print(f"   MAE:      {best_mae:.4f} kWh/m²/day")
print(f"   R2 Score: {best_r2:.4f} ({best_r2*100:.2f}% accuracy)")

# Save comparison results
comparison_df.to_csv('model_comparison_results.csv', index=False)
print(f"\nComparison results saved: 'model_comparison_results.csv'")

# Visualization 8: Model Comparison Charts
print("\n   [8/10] Model comparison charts...")
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# MAE Comparison
axes[0].bar(comparison_df['Model'], comparison_df['MAE (kWh/m²/day)'],
           color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'][:len(comparison_df)])
axes[0].set_title('Mean Absolute Error (MAE)\nLower is Better', fontweight='bold', fontsize=14)
axes[0].set_ylabel('MAE (kWh/m²/day)')
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)
for i, v in enumerate(comparison_df['MAE (kWh/m²/day)']):
    axes[0].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=10, fontweight='bold')

# RMSE Comparison
axes[1].bar(comparison_df['Model'], comparison_df['RMSE (kWh/m²/day)'],
           color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'][:len(comparison_df)])
axes[1].set_title('Root Mean Squared Error (RMSE)\nLower is Better', fontweight='bold', fontsize=14)
axes[1].set_ylabel('RMSE (kWh/m²/day)')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', alpha=0.3)
for i, v in enumerate(comparison_df['RMSE (kWh/m²/day)']):
    axes[1].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=10, fontweight='bold')

# R2 Score Comparison
axes[2].bar(comparison_df['Model'], comparison_df['R2 Score'],
           color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'][:len(comparison_df)])
axes[2].set_title('R2 Score (Accuracy)\nHigher is Better', fontweight='bold', fontsize=14)
axes[2].set_ylabel('R2 Score')
axes[2].axhline(y=0.8, color='red', linestyle='--', alpha=0.5, label='Good Threshold (0.80)')
axes[2].legend()
axes[2].tick_params(axis='x', rotation=45)
axes[2].grid(axis='y', alpha=0.3)
for i, v in enumerate(comparison_df['R2 Score']):
    axes[2].text(i, v + 0.01, f'{v:.3f}', ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('visualizations/8_model_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 9: All Model Predictions
print("   [9/10] Model predictions comparison...")
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
axes = axes.flatten()

for idx, (model_name, model_results) in enumerate(results.items()):
    if model_results is not None:
        ax = axes[idx]
        
        # Plot actual vs predicted
        ax.plot(dates_test, y_test.values, label='Actual Solar Index', 
               linewidth=2, alpha=0.8, color='blue')
        ax.plot(dates_test, model_results['predictions'], label='Predicted Solar Index',
               linewidth=2, alpha=0.8, color='red', linestyle='--')
        
        ax.set_title(f'{model_name} - Predictions vs Actual', fontweight='bold', fontsize=14)
        ax.set_xlabel('Date', fontsize=11)
        ax.set_ylabel('Solar Index (kWh/m²/day)', fontsize=11)
        ax.legend(fontsize=10)
        ax.grid(True, alpha=0.3)
        
        # Add metrics box
        metrics_text = (f"MAE: {model_results['mae']:.3f}\n"
                       f"RMSE: {model_results['rmse']:.3f}\n"
                       f"R2: {model_results['r2']:.3f}")
        ax.text(0.02, 0.98, metrics_text, transform=ax.transAxes,
               verticalalignment='top', fontsize=10,
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.7))

plt.suptitle(f'Solar Index Prediction - {LOCATION}, Mumbai\nModel Comparison', 
            fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.savefig('visualizations/9_all_predictions.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 10: Feature Importance
print("   [10/10] Feature importance analysis...")
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Random Forest Feature Importance
rf_importance = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=True).tail(15)

axes[0].barh(rf_importance['Feature'], rf_importance['Importance'], color='steelblue')
axes[0].set_xlabel('Importance Score', fontsize=12)
axes[0].set_title('Random Forest - Top 15 Important Features', fontweight='bold', fontsize=14)
axes[0].grid(axis='x', alpha=0.3)

# XGBoost Feature Importance
xgb_importance = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': xgb_model.feature_importances_
}).sort_values('Importance', ascending=True).tail(15)

axes[1].barh(xgb_importance['Feature'], xgb_importance['Importance'], color='coral')
axes[1].set_xlabel('Importance Score', fontsize=12)
axes[1].set_title('XGBoost - Top 15 Important Features', fontweight='bold', fontsize=14)
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('visualizations/10_feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nAll visualizations completed!")

# ==================== STEP 9: RESIDUAL ANALYSIS ====================
print("\n" + "="*70)
print("RESIDUAL ANALYSIS")
print("="*70)

best_model_key = best_model_name
best_predictions = results[best_model_key]['predictions']
residuals = y_test.values - best_predictions

print(f"\nResidual Statistics for {best_model_name}:")
print(f"   Mean:     {np.mean(residuals):.4f}")
print(f"   Std Dev:  {np.std(residuals):.4f}")
print(f"   Min:      {np.min(residuals):.4f}")
print(f"   Max:      {np.max(residuals):.4f}")

# Residual plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Residual vs Predicted
axes[0, 0].scatter(best_predictions, residuals, alpha=0.5, s=30)
axes[0, 0].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[0, 0].set_xlabel('Predicted Solar Index (kWh/m²/day)')
axes[0, 0].set_ylabel('Residuals')
axes[0, 0].set_title('Residuals vs Predicted Values', fontweight='bold', fontsize=14)
axes[0, 0].grid(True, alpha=0.3)

# Residual distribution
axes[0, 1].hist(residuals, bins=30, edgecolor='black', alpha=0.7, color='skyblue')
axes[0, 1].axvline(x=0, color='red', linestyle='--', linewidth=2)
axes[0, 1].set_xlabel('Residuals')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Residual Distribution', fontweight='bold', fontsize=14)
axes[0, 1].grid(axis='y', alpha=0.3)

# Q-Q plot
from scipy import stats
stats.probplot(residuals, dist="norm", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot', fontweight='bold', fontsize=14)
axes[1, 0].grid(True, alpha=0.3)

# Residuals over time
axes[1, 1].plot(dates_test, residuals, linewidth=1, alpha=0.7, color='purple')
axes[1, 1].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Residuals')
axes[1, 1].set_title('Residuals Over Time', fontweight='bold', fontsize=14)
axes[1, 1].grid(True, alpha=0.3)

plt.suptitle(f'Residual Analysis - {best_model_name} Model', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('visualizations/11_residual_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== STEP 10: FUTURE PREDICTIONS ====================
print("\n" + "="*70)
print("GENERATING FUTURE PREDICTIONS (30 DAYS)")
print("="*70)

# Use best model
if best_model_name in ['Random Forest', 'XGBoost']:
    best_model = results[best_model_key]['model']
    
    # Create future dates
    last_date = df_clean['Date'].max()
    future_dates = pd.date_range(start=last_date + timedelta(days=1), periods=30, freq='D')
    
    print(f"\nForecast Period: {future_dates[0]} to {future_dates[-1]}")
    
    # Prepare future features (using recent averages and last known values)
    future_data = []
    future_predictions_list = []  # Initialize list to store predictions as we generate them
    
    for i, date in enumerate(future_dates):
        # Use rolling averages from recent data
        recent_data = df_clean.tail(30)
        
        # For lag features, use either historical data or previously predicted values
        if i == 0:
            solar_lag_1 = df_clean['Solar_Index'].iloc[-1]
            solar_lag_2 = df_clean['Solar_Index'].iloc[-2]
            solar_lag_3 = df_clean['Solar_Index'].iloc[-3]
        elif i == 1:
            solar_lag_1 = future_predictions_list[0]
            solar_lag_2 = df_clean['Solar_Index'].iloc[-1]
            solar_lag_3 = df_clean['Solar_Index'].iloc[-2]
        elif i == 2:
            solar_lag_1 = future_predictions_list[1]
            solar_lag_2 = future_predictions_list[0]
            solar_lag_3 = df_clean['Solar_Index'].iloc[-1]
        else:
            solar_lag_1 = future_predictions_list[i-1]
            solar_lag_2 = future_predictions_list[i-2]
            solar_lag_3 = future_predictions_list[i-3]
        
        # For lag_7, always use historical data since we don't have 7 future predictions yet
        if i < 7:
            solar_lag_7 = df_clean['Solar_Index'].iloc[-7+i]
        else:
            solar_lag_7 = future_predictions_list[i-7]
        
        future_features = {
            'Temperature': recent_data['Temperature'].mean(),
            'Humidity': recent_data['Humidity'].mean(),
            'Wind_Speed': recent_data['Wind_Speed'].mean(),
            'Precipitation': recent_data['Precipitation'].mean(),
            'Month': date.month,
            'DayOfYear': date.dayofyear,
            'DayOfWeek': date.dayofweek,
            'Solar_Lag_1': solar_lag_1,
            'Solar_Lag_2': solar_lag_2,
            'Solar_Lag_3': solar_lag_3,
            'Solar_Lag_7': solar_lag_7,
            'Solar_Rolling_Mean_3': recent_data['Solar_Index'].tail(3).mean(),
            'Solar_Rolling_Mean_7': recent_data['Solar_Index'].tail(7).mean(),
            'Solar_Rolling_Mean_14': recent_data['Solar_Index'].tail(14).mean(),
            'Solar_Rolling_Std_7': recent_data['Solar_Index'].tail(7).std(),
            'Solar_EWMA_7': recent_data['Solar_Index'].ewm(span=7).mean().iloc[-1]
        }
        future_data.append(future_features)
        
        # Create temporary dataframe for current prediction
        temp_future_df = pd.DataFrame([future_features])
        
        # Scale features
        temp_future_scaled = scaler.transform(temp_future_df)
        
        # Predict current day
        current_prediction = best_model.predict(temp_future_scaled)[0]
        future_predictions_list.append(current_prediction)
    
    # Convert to numpy array for consistency
    future_predictions = np.array(future_predictions_list)
    
    # Create results dataframe
    forecast_df = pd.DataFrame({
        'Date': future_dates,
        'Predicted_Solar_Index': future_predictions,
        'Day_of_Week': [date.strftime('%A') for date in future_dates],
        'Season': [get_season(date.month) for date in future_dates]
    })
    
    # Add confidence intervals (approximation)
    forecast_df['Lower_Bound'] = forecast_df['Predicted_Solar_Index'] - 2 * results[best_model_key]['rmse']
    forecast_df['Upper_Bound'] = forecast_df['Predicted_Solar_Index'] + 2 * results[best_model_key]['rmse']
    
    # Ensure bounds are non-negative
    forecast_df['Lower_Bound'] = forecast_df['Lower_Bound'].clip(lower=0)
    
    print("\n30-Day Solar Index Forecast:")
    print(forecast_df.to_string(index=False))
    
    # Save forecast
    forecast_df.to_csv('solar_index_30day_forecast.csv', index=False)
    print(f"\nForecast saved: 'solar_index_30day_forecast.csv'")
    
    # Visualize forecast
    fig, ax = plt.subplots(figsize=(16, 8))
    
    # Historical data (last 90 days)
    historical = df_clean.tail(90)
    ax.plot(historical['Date'], historical['Solar_Index'], 
           label='Historical Solar Index', linewidth=2, alpha=0.8, color='blue')
    
    # Forecast
    ax.plot(forecast_df['Date'], forecast_df['Predicted_Solar_Index'],
           label='30-Day Forecast', linewidth=2.5, color='red', marker='o', markersize=6)
    
    # Confidence interval
    ax.fill_between(forecast_df['Date'], 
                    forecast_df['Lower_Bound'], 
                    forecast_df['Upper_Bound'],
                    alpha=0.3, color='red', label='95% Confidence Interval')
    
    # Today's line
    ax.axvline(x=last_date, color='green', linestyle='--', linewidth=2, 
              alpha=0.7, label='Today')
    
    ax.set_title(f'Solar Index Forecast - {LOCATION}, Mumbai (Next 30 Days)\nModel: {best_model_name}',
                fontsize=16, fontweight='bold')
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Solar Index (kWh/m²/day)', fontsize=12)
    ax.legend(fontsize=11)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('visualizations/12_future_forecast.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("   Forecast visualization saved")
    
    # Summary statistics
    print("\nForecast Summary:")
    print(f"   Average Predicted Solar Index: {forecast_df['Predicted_Solar_Index'].mean():.2f} kWh/m²/day")
    print(f"   Maximum Predicted Solar Index: {forecast_df['Predicted_Solar_Index'].max():.2f} kWh/m²/day")
    print(f"   Minimum Predicted Solar Index: {forecast_df['Predicted_Solar_Index'].min():.2f} kWh/m²/day")
    print(f"   Standard Deviation: {forecast_df['Predicted_Solar_Index'].std():.2f} kWh/m²/day")

elif best_model_name == 'Prophet':
    print("Using Prophet for future predictions...")
    # Prophet already has built-in future prediction capability
    future = prophet_model.make_future_dataframe(periods=30)
    forecast = prophet_model.predict(future)
    
    # Extract the future predictions
    future_forecast = forecast[forecast['ds'] > df_clean['Date'].max()].tail(30)
    
    forecast_df = pd.DataFrame({
        'Date': future_forecast['ds'],
        'Predicted_Solar_Index': future_forecast['yhat'],
        'Lower_Bound': future_forecast['yhat_lower'],
        'Upper_Bound': future_forecast['yhat_upper'],
        'Day_of_Week': [date.strftime('%A') for date in future_forecast['ds']],
        'Season': [get_season(date.month) for date in future_forecast['ds']]
    })
    
    print("\n30-Day Solar Index Forecast (Prophet):")
    print(forecast_df.to_string(index=False))
    
    # Save forecast
    forecast_df.to_csv('solar_index_30day_forecast.csv', index=False)
    print(f"\nForecast saved: 'solar_index_30day_forecast.csv'")

else:
    print(f"Future prediction not implemented for {best_model_name} model")
# ==================== STEP 11: KEY INSIGHTS & RECOMMENDATIONS ====================
print("\n" + "="*70)
print("KEY INSIGHTS & FINDINGS")
print("="*70)

print(f"\nSOLAR INDEX ANALYSIS - {LOCATION}, Mumbai")
print("-" * 70)

# Seasonal insights
seasonal_stats = df_clean.groupby('Season')['Solar_Index'].agg(['mean', 'std', 'min', 'max'])
seasonal_stats = seasonal_stats.reindex(['Winter', 'Summer', 'Monsoon', 'Post-Monsoon'])

print("\n1. SEASONAL PATTERNS:")
for season in seasonal_stats.index:
    stats = seasonal_stats.loc[season]
    print(f"   {season:15} -> Avg: {stats['mean']:.2f} kWh/m²/day  "
          f"(Range: {stats['min']:.2f} - {stats['max']:.2f})")

best_season = seasonal_stats['mean'].idxmax()
worst_season = seasonal_stats['mean'].idxmin()
print(f"\n   Best Season for Solar:  {best_season} ({seasonal_stats.loc[best_season, 'mean']:.2f} kWh/m²/day)")
print(f"   Worst Season for Solar: {worst_season} ({seasonal_stats.loc[worst_season, 'mean']:.2f} kWh/m²/day)")

# Monthly insights
monthly_stats = df_clean.groupby('Month')['Solar_Index'].mean()
best_month = monthly_stats.idxmax()
worst_month = monthly_stats.idxmin()
month_names = {1:'January', 2:'February', 3:'March', 4:'April', 5:'May', 6:'June',
              7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}

print(f"\n2. MONTHLY PATTERNS:")
print(f"   Best Month:  {month_names[best_month]} ({monthly_stats[best_month]:.2f} kWh/m²/day)")
print(f"   Worst Month: {month_names[worst_month]} ({monthly_stats[worst_month]:.2f} kWh/m²/day)")

# Weather correlations
print(f"\n3. WEATHER IMPACT:")
temp_corr = df_clean[['Solar_Index', 'Temperature']].corr().iloc[0, 1]
hum_corr = df_clean[['Solar_Index', 'Humidity']].corr().iloc[0, 1]
print(f"   Temperature: {'Positive' if temp_corr > 0 else 'Negative'} correlation (r={temp_corr:.3f})")
print(f"   Humidity:    {'Positive' if hum_corr > 0 else 'Negative'} correlation (r={hum_corr:.3f})")

# Model performance
print(f"\n4. PREDICTION ACCURACY:")
print(f"   Best Model: {best_model_name}")
print(f"   Accuracy: {best_r2*100:.2f}%")
print(f"   Average Error: +-{best_mae:.3f} kWh/m²/day")

# Top features
print(f"\n5. KEY PREDICTIVE FACTORS:")
if best_model_name in ['Random Forest', 'XGBoost']:
    importance = results[best_model_key]['model'].feature_importances_
    top_features = pd.DataFrame({
        'Feature': feature_cols,
        'Importance': importance
    }).sort_values('Importance', ascending=False).head(5)
    
    for idx, row in top_features.iterrows():
        print(f"   {row['Feature']:30} -> {row['Importance']*100:.1f}% importance")

# ==================== STEP 12: RECOMMENDATIONS ====================
print("\n" + "="*70)
print("RECOMMENDATIONS FOR SOLAR ENERGY PLANNING")
print("="*70)

print("\n1. OPTIMAL INSTALLATION TIMING:")
print(f"   Install solar panels before {month_names[best_month]} for maximum benefit")
print(f"   Plan maintenance during {worst_season} season (lowest solar index)")

print("\n2. EXPECTED SOLAR PERFORMANCE:")
annual_avg = df_clean['Solar_Index'].mean()
print(f"   Average daily solar index: {annual_avg:.2f} kWh/m²/day")
print(f"   Annual solar potential: ~{annual_avg * 365:.0f} kWh/m²/year")

print("\n3. ENERGY PLANNING:")
print(f"   Summer months can generate {(seasonal_stats.loc['Summer', 'mean']/annual_avg - 1)*100:+.1f}% above average")
print(f"   Monsoon months generate {(seasonal_stats.loc['Monsoon', 'mean']/annual_avg - 1)*100:+.1f}% below average")
print(f"   Plan backup energy sources for monsoon period")

print("\n4. MODEL DEPLOYMENT:")
print(f"   Use {best_model_name} for daily predictions")
print(f"   Expected prediction error: +-{best_mae:.2f} kWh/m²/day")
print(f"   Retrain model quarterly with new data")

# ==================== STEP 13: PROJECT SUMMARY ====================
print("\n" + "="*70)
print("PROJECT SUMMARY")
print("="*70)

print(f"\nLocation: {LOCATION}, Mumbai ({locations[LOCATION]['lat']}N, {locations[LOCATION]['lon']}E)")
print(f"Analysis Period: {df_clean['Date'].min()} to {df_clean['Date'].max()}")
print(f"Total Days Analyzed: {len(df_clean)}")
print(f"Models Trained: 4 (ARIMA, Random Forest, XGBoost, Prophet)")
print(f"Best Model: {best_model_name} (R2={best_r2:.4f})")
print(f"Average Solar Index: {annual_avg:.2f} kWh/m²/day")
print(f"Forecast Period: Next 30 days")

print("\nOUTPUT FILES GENERATED:")
print(f"   1. solar_index_data_{LOCATION}.csv")
print(f"   2. solar_index_data_{LOCATION}_processed.csv")
print(f"   3. model_comparison_results.csv")
print(f"   4. solar_index_30day_forecast.csv")
print(f"   5. visualizations/ (12 plots)")

print("\nVISUALIZATIONS CREATED:")
viz_files = [
    "1_solar_index_timeseries.png",
    "2_seasonal_analysis.png",
    "3_correlation_matrix.png",
    "4_year_comparison.png",
    "5_weather_impact.png",
    "6_monthly_heatmap.png",
    "7_decomposition.png",
    "8_model_comparison.png",
    "9_all_predictions.png",
    "10_feature_importance.png",
    "11_residual_analysis.png",
    "12_future_forecast.png"
]
for i, viz in enumerate(viz_files, 1):
    print(f"   {i:2}. {viz}")

print("\n" + "="*70)
print("PROJECT COMPLETED SUCCESSFULLY!")
print("="*70)

print(f"""

****************************************************************
                    ANALYSIS COMPLETE!                   
                                                                
  Solar Index Prediction System is ready for deployment        
  All data, models, and visualizations have been generated     
                                                                
  Next Steps:                                                   
  1. Review the visualizations in 'visualizations/' folder     
  2. Check model comparison results in CSV files               
  3. Use forecast data for solar planning                      
  4. Upload to GitHub for project submission                   
                                                                
  Thank you for using this Solar Index Prediction System!   
****************************************************************
""")

# Print file locations
print(f"\nAll files saved in: {os.getcwd()}")
print(f"Ready for GitHub upload and project submission!")
print(f"\nData Source: NASA POWER API (https://power.larc.nasa.gov/)")
print(f"For questions or support, check the project README.md\n")
```
